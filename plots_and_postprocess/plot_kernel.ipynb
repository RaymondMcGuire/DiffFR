```python
import matplotlib.pyplot as plt
import numpy as np

h = 0.25

def W(x):
    q = x / h
    s = 4 / (3 * h)
    if q >=0 and q <= 0.5:
        return s * (6 * (q ** 3 - q ** 2) + 1)
    elif q >=0.5 and q <= 1: 
        return s * 2 * (1 - q) ** 3
    else:
        return 0 
        
def gradW(x):
    q = x / h
    s = 4 / (3 * h)
    if q >=0 and q <= 0.5:
        return s * (6/h * (3 * q**2 - 2 * q))
    elif q >=0.5 and q <= 1: 
        return -s * 6/h * (1 - q) ** 2
    else:
        return 0 

def gradGradW(x):
    q = x / h
    s = 4 / (3 * h)
    if q >=0 and q <= 0.5:
        return s *6/h/h* ((3 *q**2 - 2 * q) + 6 * q - 2)
    elif q >=0.5 and q <= 1: 
        return s * 6 /h/h * (- (1 - q) ** 2 + 2*(1-q))
    else:
        return 0 

X = np.linspace(0, 0.26, 10000)

plt.figure(dpi=200)
plt.plot(X, [0. for x in X], 'black')
# plt.plot(X, [np.log(np.abs(W(x))) for x in X], label='$W$')
# plt.plot(X, [np.log(np.abs(gradW(x))) for x in X], label='$ \\nabla W$')
# plt.plot(X, [np.log(np.abs(gradGradW(x))) for x in X], label='$ \\nabla^2 W$')
plt.plot(X, [W(x) for x in X], label='$W$')
plt.plot(X, [0.1 * gradW(x) for x in X], label='$0.1 * \\nabla W$')
plt.plot(X, [0.005 * gradGradW(x) for x in X], label='$0.005 * \\nabla^2 W$')

plt.xlabel('distance', fontsize=14)
plt.ylabel('value', fontsize=14)
# plt.ylabel('')
plt.tick_params(axis='x', labelsize=14)
plt.tick_params(axis='y', labelsize=14)
plt.grid(True)
plt.xticks(np.arange(0.0, 0.26, 0.05))
plt.yticks(np.arange(-5.0, 7.0, 1.0))
plt.legend(fontsize=14)
plt.show()

```
